```python
from dotenv import load_dotenv
import os

load_dotenv()

db_user = os.getenv('db_user')
db_password = os.getenv('db_password')
db_host = os.getenv('db_host')
db_port = os.getenv('db_port')
db_name = os.getenv('db_name')
```

```python
print(db_user)
```

```python
database_url = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
```

```python
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session

engine = create_engine(
    database_url,
    connect_args={"options": "-csearch_path=endeavour"}
)

Base = automap_base()

# ✅ modern reflection
Base.prepare(autoload_with=engine,schema="endeavour")

print(Base.classes.keys())
```

```python
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session

engine = create_engine(database_url)

metadata = MetaData(schema="endeavour")  # 🔥 IMPORTANT

Base = automap_base(metadata=metadata)
Base.prepare(autoload_with=engine)

print(Base.classes.keys())  # 👀 debug
```

```python
from sqlalchemy import inspect

inspector = inspect(engine)

print(inspector.get_table_names(schema="public"))
print(inspector.get_table_names(schema="endeavour"))
```
