```python
import sqlite3
import csv

conn = sqlite3.connect("sales.db")
cursor = conn.cursor()

with open("../csv files/sales.csv", "r") as file:
    dr = csv.DictReader(file)
    columns = dr.fieldnames

    # create table dynamically
    cursor.execute(f"CREATE TABLE IF NOT EXISTS sales_table ({','.join(columns)})")

    for row in dr:
        cursor.execute(
            f"INSERT INTO sales_table VALUES ({','.join(['?']*len(columns))})",
            list(row.values())
        )

conn.commit()
```

```python
import pandas as pd
```

```python
df=pd.read_sql('select * from sales_table', conn)
```
