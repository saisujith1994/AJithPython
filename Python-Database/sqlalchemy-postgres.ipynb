```python
from dotenv import load_dotenv
import os

load_dotenv()

db_user = os.getenv('db_user')
db_password = os.getenv('db_password')
db_host = os.getenv('db_host')
db_port = os.getenv('db_port')
db_name = os.getenv('db_name')
```

```python
database_url = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
engine = create_engine(database_url)
```

```python
metadata = MetaData()
```

```python
rowdies = Table('rowdies', metadata,
                 Column('id', Integer, primary_key=True),
                 Column('name', String),
                 Column('age', Integer),
                 Column('height', Float)
                )
```

```python
metadata.create_all(engine)
```

```python
with engine.connect() as connection:
    connection.execute(
        rowdies.insert(),
        [
            {'id': 4, 'name': 'Alex', 'age': 33, 'height': 6.5},
            {'id': 5, 'name': 'Bobby', 'age': 18, 'height': 5.0},
            {'id': 6, 'name': 'xavier', 'age': 55, 'height': 5.8}
        ]
    )
    connection.commit()
```

Query data

```python
with engine.connect() as connection:
    result= connection.execute(select(rowdies))
    for row in result:
        print(row)
```
